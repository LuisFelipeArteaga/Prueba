#### Q22. ¿Cuáles son los tipos de sesgos que pueden ocurrir durante el muestreo? (What Are the Types of Biases That Can Occur During Sampling?)

El sesgo de muestreo ocurre cuando la muestra de datos en una investigación sistemática no representa con precisión lo que se puede obtener en el entorno de la investigación. Cuando recopila datos de manera que algunos miembros de la población prevista tengan una probabilidad de muestreo menor o mayor que otros, el resultado es un sesgo de muestreo. 

Tipos de sesgo de muestreo:

    Self-Selection: ocurre cuando los individuos con características específicas se seleccionan a sí mismos en la muestra de investigación.
    
    Undercoverage:cuando algunas de las variables de la población están mal representadas o no están representadas en la muestra del estudio.
    
    Non-response Bias:es la incapacidad de una parte de la población de su estudio para participar en el estudio debido a un factor que los hace diferir mucho del resto de la población.
    
    Supervivencia:ocurre cuando ignora las variables de investigación que no lograron superar un proceso de selección natural o antinatural mientras presta atención a las variables que sí lo hicieron
    
    Healthy User: es común en estudios médicos y epidemiológicos. El sesgo de muestreo de usuarios saludables simplemente significa que el tipo de personas que se ofrecen como voluntarias para la investigación médica y los ensayos clínicos a menudo está muy lejos de lo que se puede obtener en la población general
    
    Pre-screening or Advertising Bias: ocurre cuando el proceso de selección implementado en un estudio da como resultado una muestra que es una representación deficiente de la población.

#### Q25. Explique cómo funciona una curva ROC. (Explain how a ROC curve works?)

![Image](https://miro.medium.com/max/500/0*vLy0-cJN-uilKPBM.png)

ROC) es un diagrama gráfico que se utiliza para mostrar la capacidad de diagnóstico de los clasificadores binarios. Una curva ROC se construye trazando la tasa de verdaderos positivos (TPR) contra la tasa de falsos positivos (FPR). La tasa de verdaderos positivos es la proporción de observaciones que se predijo correctamente que serían positivas de todas las observaciones positivas (TP / (TP + FN)). De manera similar, la tasa de falsos positivos es la proporción de observaciones que se predice incorrectamente como positivas de todas las observaciones negativas (FP / (TN + FP)).

La curva ROC muestra el compromiso entre sensibilidad (o TPR) y especificidad (1 - FPR). Los clasificadores que dan curvas más cercanas a la esquina superior izquierda indican un mejor rendimiento. Como línea de base, se espera que un clasificador aleatorio dé puntos a lo largo de la diagonal (FPR = TPR). Cuanto más se acerque la curva a la diagonal de 45 grados del espacio ROC, menos precisa será la prueba.

Para comparar diferentes clasificadores, puede ser útil resumir el desempeño de cada clasificador en una sola medida. Un método común es calcular el área bajo la curva ROC, que se abrevia como AUC.

#### Q38. ¿Puede explicar la diferencia entre un conjunto de validación y un conjunto de prueba? (Can you explain the difference between a Validation Set and a Test Set?)

#### Q48. Explique el algoritmo del árbol de decisión en detalle. (Explain Decision Tree algorithm in detail.)

El árbol de decisiones es una técnica de aprendizaje supervisado que se puede utilizar tanto para problemas de clasificación como de regresión, pero sobre todo se prefiere para resolver problemas de clasificación. Es un clasificador estructurado en árbol, donde los nodos internos representan las características de un conjunto de datos, las ramas representan las reglas de decisión y cada nodo hoja representa el resultado.

En un árbol de decisión, hay dos nodos, que son el nodo de decisión y el nodo hoja. Los nodos de decisión se utilizan para tomar cualquier decisión y tienen múltiples ramas, mientras que los nodos hoja son el resultado de esas decisiones y no contienen más ramas.

Las decisiones o la prueba se realizan sobre la base de las características del conjunto de datos dado.

Se le llama árbol de decisión porque, similar a un árbol, comienza con el nodo raíz, que se expande en más ramas y construye una estructura similar a un árbol.

Para construir un árbol, usamos el algoritmo CART, que significa algoritmo de árbol de clasificación y regresión.

Un árbol de decisiones simplemente hace una pregunta y, según la respuesta (Sí / No), divide el árbol en subárboles.

El siguiente diagrama explica la estructura general de un árbol de decisiones:

![Image](https://static.javatpoint.com/tutorial/machine-learning/images/decision-tree-classification-algorithm.png)

**Nodo raíz**: el nodo raíz es desde donde comienza el árbol de decisiones. Representa el conjunto de datos completo, que además se divide en dos o más conjuntos homogéneos.

**Nodo de hoja**: los nodos de hoja son el nodo de salida final y el árbol no se puede segregar más después de obtener un nodo de hoja.

**Dividir**: Dividir es el proceso de dividir el nodo de decisión / nodo raíz en subnodos de acuerdo con las condiciones dadas.

**Rama / Subárbol**: Un árbol formado al dividir el árbol.

**Poda**: la poda es el proceso de eliminar las ramas no deseadas del árbol.

**Nodo principal / secundario**: el nodo raíz del árbol se denomina nodo principal y los otros nodos se denominan nodos secundarios.

En un árbol de decisión, para predecir la clase del conjunto de datos dado, el algoritmo comienza desde el nodo raíz del árbol. Este algoritmo compara los valores del atributo raíz con el atributo de registro (conjunto de datos real) y, según la comparación, sigue la rama y salta al siguiente nodo.

Para el siguiente nodo, el algoritmo vuelve a comparar el valor del atributo con los otros subnodos y avanza. Continúa el proceso hasta que llega al nodo de la hoja del árbol. El proceso completo se puede comprender mejor utilizando el siguiente algoritmo:

**Paso 1:** Comience el árbol con el nodo raíz, dice S, que contiene el conjunto de datos completo.

**Paso 2:** Encuentre el mejor atributo en el conjunto de datos usando la Medida de selección de atributos (ASM).

**Paso 3:** Divida la S en subconjuntos que contengan valores posibles para los mejores atributos.

**Paso 4:** Genere el nodo del árbol de decisión, que contiene el mejor atributo.

**Paso 5:** Cree de forma recursiva nuevos árboles de decisión utilizando los subconjuntos del conjunto de datos creado en el paso -3. Continúe este proceso hasta que se alcance una etapa en la que no pueda clasificar los nodos más y se llame al nodo final como un nodo hoja.

El método multivariado intenta resolver eso construyendo un modelo usando todos los datos disponibles y luego limpiando aquellas instancias con errores por encima de un valor dado.

**Error de Minkowski**

A diferencia de los métodos univariante y multivariante, no detecta ni limpia los valores atípicos. En cambio, reduce el impacto que tendrán los valores atípicos en el modelo.

El error de Minkowski es un índice de pérdida que es más insensible a los valores atípicos que el error estándar de suma cuadrada. El error de suma al cuadrado eleva cada error de instancia al cuadrado, lo que hace una contribución demasiado grande de valores atípicos al error total. El error de Minkowski lo resuelve elevando cada error de instancia a un número menor que 2, por ejemplo 1,5. Esto reduce la contribución de los valores atípicos al error total. Por ejemplo, si un valor atípico tiene un error de 10, el error al cuadrado para esa instancia será 100, mientras que el error de Minkowski será 31,62.

#### Q60. ¿Cómo definiría el número de clústeres en un algoritmo de agrupamiento? (How will you define the number of clusters in a clustering algorithm?)

**Método del codo**
La idea básica de los algoritmos de clustering es la minimización de la varianza intra-cluster y la maximización de la varianza inter-cluster. Es decir, queremos que cada observación se encuentre muy cerca a las de su mismo grupo y los grupos lo más lejos posible entre ellos.

El método del codo utiliza la distancia media de las observaciones a su centroide. Es decir, se fija en las distancias intra-cluster. Cuanto más grande es el número de clusters k, la varianza intra-cluster tiende a disminuir. Cuanto menor es la distancia intra-cluster mejor, ya que significa que los clústers son más compactos. El método del codo busca el valor k que satisfaga que un incremento de k, no mejore sustancialmente la distancia media intra-cluster

![Image](https://i2.wp.com/www.aprendemachinelearning.com/wp-content/uploads/2018/03/ejemplo-elbow.png?resize=768%2C422)

**Análisis de la silueta:**
El análisis de la silueta mide la calidad del agrupamiento o clustering. Mide la distancia de separación entre los clústers. Nos indica como de cerca está cada punto de un clúster a puntos de los clústers vecinos. Esta medida de distancia se encuentra en el rango $[-1, 1]$. Un valor alto indica un buen clustering.

Los coeficientes de silueta cercanos a +1 indican que la observación se encuentra lejos de los clústers vecinos. Un valor del coeficiente de 0 indica que la observación está muy cerca o en la frontera de decisión entre dos clústers. Valores negativos indican que esas muestras quizás estén asignadas al clúster erróneo.

El método de la silueta calcula la media de los coeficientes de silueta de todas las observaciones para diferentes valores de k. El número óptimo de clústers k es aquel que maximiza la media de los coeficientes de silueta para un rango de valores de k.

El coeficiente de la silueta es calculado como:

$s = \frac{b-a}{max(a,b)}$

siendo a la distancia media intra-clúster y b la distancia media a las observaciones del clúster más cercano.

#### Q65. ¿Qué técnica de validación cruzada usaría en un conjunto de datos de series de tiempo? (What cross-validation technique would you use on a time series data set?)

El método que se puede utilizar para la validación cruzada del modelo de serie temporal es la validación cruzada de forma continua. Funciona agregando márgenes en dos posiciones. La primera mitad de los datos (divididos temporalmente) se asigna al conjunto de entrenamiento y la segunda mitad se convierte en el conjunto de prueba. #### Q79. ¿Cuál es la diferencia entre época, lote e iteración en el aprendizaje profundo? What Is the Difference Between Epoch, Batch, and Iteration in Deep Learning?

**Época**
La red neuronal aprende los patrones de los datos de entrada leyendo el conjunto de datos de entrada y aplicando diferentes cálculos en él. Pero la red neuronal no lo hace solo una vez, aprende, una y otra vez, utilizando el conjunto de datos de entrada y los resultados obtenidos de las pruebas anteriores.

Cada sendero para aprender del conjunto de datos de entrada se denomina época.

Entonces, una época se refiere a un ciclo a través del conjunto de datos de entrenamiento completo. Por lo general, entrenar una red neuronal lleva más de unas pocas épocas. Aumentar el número de épocas no siempre significa que la red dará mejores resultados.

Entonces, básicamente, por ensayo y error, elegimos varias épocas en las que los resultados siguen siendo los mismos después de unos pocos ciclos.

**Iteración**
Para cada época completa, tenemos varias iteraciones. La iteración es el número de lotes o pasos a través de paquetes particionados de los datos de entrenamiento, necesarios para completar una época.

**Lote**
Lote es el número de muestras de entrenamiento o ejemplos en una iteración. Cuanto mayor sea el tamaño del lote, más espacio de memoria necesitaremos.

#### Q80. ¿Cuáles son las diferentes capas en CNN? (What Are the Different Layers on CNN?)

Convolutional Layer

Esta capa es la primera capa que se utiliza para extraer las diversas características de las imágenes de entrada. En esta capa, la operación matemática de convolución se realiza entre la imagen de entrada y un filtro de un tamaño particular MxM. Al deslizar el filtro sobre la imagen de entrada, el producto escalar se toma entre el filtro y las partes de la imagen de entrada con respecto al tamaño del filtro (MxM).

La salida se denomina mapa de características que nos brinda información sobre la imagen, como las esquinas y los bordes. Posteriormente, este mapa de características se envía a otras capas para conocer otras características de la imagen de entrada.

Pooling Layer

En la mayoría de los casos, una capa convolucional va seguida de una capa de agrupación. El objetivo principal de esta capa es disminuir el tamaño del mapa de características convolucionado para reducir los costos computacionales. Esto se realiza disminuyendo las conexiones entre capas y opera de forma independiente en cada mapa de características. Según el método utilizado, existen varios tipos de operaciones de agrupación.

En Max Pooling, el elemento más grande se toma del mapa de características. La agrupación promedio calcula el promedio de los elementos en una sección de imagen de tamaño predefinido. La suma total de los elementos de la sección predefinida se calcula en Sum Pooling. La capa de agrupación suele servir como puente entre la capa convolucional y la capa FC.

Fully Connected Layer

La capa Fully Connected (FC) consta de los pesos y sesgos junto con las neuronas y se utiliza para conectar las neuronas entre dos capas diferentes. Estas capas generalmente se colocan antes de la capa de salida y forman las últimas capas de una arquitectura CNN.

En esto, la imagen de entrada de las capas anteriores se aplana y se alimenta a la capa FC. El vector aplanado luego se somete a unas pocas capas FC más donde suelen tener lugar las operaciones de las funciones matemáticas. En esta etapa comienza a tener lugar el proceso de clasificación.

#### Q85. Explica el descenso por gradiente. (Gradient Descent)

El descenso de gradientes es un algoritmo de optimización que se usa comúnmente para entrenar  modelos de aprendizaje automático y redes neuronales . Los datos de entrenamiento ayudan a estos modelos a aprender con el tiempo, y la función de costo dentro del descenso del gradiente actúa específicamente como un barómetro, midiendo su precisión con cada iteración de actualizaciones de parámetros. Hasta que la función sea cercana o igual a cero, el modelo continuará ajustando sus parámetros para producir el menor error posible. Una vez que los modelos de aprendizaje automático están optimizados para su precisión, pueden convertirse en herramientas poderosas para aplicaciones de inteligencia artificial (IA) y ciencias de la computación. 

El punto de partida es solo un punto arbitrario para que podamos evaluar el rendimiento. La pendiente informará las actualizaciones de los parámetros, es decir, los pesos y el sesgo. La pendiente en el punto de partida será más pronunciada, pero a medida que se generen nuevos parámetros, la pendiente debe reducirse gradualmente hasta alcanzar el punto más bajo de la curva, conocido como punto de convergencia.   

De manera similar a encontrar la línea de mejor ajuste en la regresión lineal, el objetivo del descenso de gradiente es minimizar la función de costo, o el error entre y pronosticado y real. Para hacer esto, se requieren dos puntos de datos: una dirección y una tasa de aprendizaje. Estos factores determinan los cálculos de derivadas parciales de iteraciones futuras, lo que le permite llegar gradualmente al mínimo local o global (es decir, punto de convergencia). #### Q97. ¿Cuál es la diferencia entre el descenso de gradiente por lotes y el descenso de gradiente estocástico? (Batch Gradient Descent and Stochastic Gradient Descent)

**Batch Gradient Descent**

En Batch Gradient Descent, todos los datos de entrenamiento se tienen en cuenta para dar un solo paso. Tomamos el promedio de los gradientes de todos los ejemplos de entrenamiento y luego usamos ese gradiente medio para actualizar nuestros parámetros. Así que eso es solo un paso de descenso de gradiente en una época.
El descenso de gradiente por lotes es ideal para múltiples errores convexos o relativamente suaves. En este caso, nos movemos algo directamente hacia una solución óptima.

**Stochastic Gradient Descent**
 
 En Batch Gradient Descent estábamos considerando todos los ejemplos para cada paso de Gradient Descent. Pero, ¿qué pasa si nuestro conjunto de datos es muy grande? Los modelos de aprendizaje profundo anhelan datos. Cuantos más datos, más posibilidades hay de que un modelo sea bueno. Supongamos que nuestro conjunto de datos tiene 5 millones de ejemplos, entonces, solo para dar un paso, el modelo tendrá que calcular los gradientes de los 5 millones de ejemplos. Esta no parece una forma eficaz. Para abordar este problema tenemos el Descenso de gradiente estocástico. En Stochastic Gradient Descent (SGD), consideramos solo un ejemplo a la vez para dar un solo paso. Realizamos los siguientes pasos en una época para SGD:

1-Toma un ejemplo

2-Aliméntelo a la red neuronal

3-Calcular su gradiente

4-Utilice el gradiente que calculamos en el paso 3 para actualizar los pesos.

5-Repite los pasos 1 a 4 para todos los ejemplos del conjunto de datos de entrenamiento.

Dado que estamos considerando solo un ejemplo a la vez, el costo fluctuará con los ejemplos de capacitación y no necesariamente disminuirá. Pero a la larga, verá que el costo disminuye con las fluctuaciones.

Cada nodo toma 0 o más tensores como entrada y produce un tensor como salida. Permítanme darles un ejemplo de un gráfico computacional simple que consta de tres nodos: a , b y c, como se muestra a continuación:

![image info](https://ichi.pro/assets/images/max/724/1*aEtCzlfgnskczEon1tZW0w.png)

Los nodos constantes se utilizan para almacenar valores constantes ya que no requiere entrada, pero produce los valores almacenados como salida. En el ejemplo anterior, a y b son nodos constantes con valores 5 y 6 respectivamente.
El nodo c representa la operación de multiplicar el nodo constante a por b. Por lo tanto, ejecutar el nodo c resultará en la multiplicación del nodo constante ay b.


#### Q101. ¿Qué es una Red Adversarial Generativa? (Generative Adversarial Network)

Los modelos generativos están basados en redes neuronales profundas. En los diagramas vamos a utilizar la notación «DL» para indicar Deep Learning, que significa red neuronal profunda. Las redes neuronales reciben datos de entrada y produce resultados como salida.

![image info](https://www.iartificial.net/wp-content/uploads/2019/03/DL_entrada_salida.jpg)

Los modelos generativos usan 2 redes neuronales profundas. Estas dos redes son adversarias, es decir, «juegan» un juego de suma cero donde lo que una red gana, la otra pierde. Me explico. Voy a explicarlo con el ejemplo de la generación de fotos. Imagínate que queremos generar fotos de caras de personas.

La tarea del Discriminador será decir si una cara es auténtica o falsa
La tarea del Generador será la de crear fotos de caras que parezcan auténticas

![image info](https://www.iartificial.net/wp-content/uploads/2019/03/redes-generativas-adversarias-1024x291.webp)


Para entrenar esta arquitectura de redes neuronales necesitamos tener un conjunto de datos reales. En el ejemplo de las caras de personas, deberemos tener muchas fotos de personas reales. De esta forma, cuando le demos una foto real al discriminador, sabremos que la respuesta correcta es «real». Y cuando le demos una foto creada por el generador, la respuesta correcta es «falso».

Al principio, es muy fácil para el discriminador acertar cuando dice cuáles fotos son reales. Esto es así porque al principio, las fotos creadas por la red neuronal generadora parecen aleatorias. Sin embargo, a medida que el generador mejora, la tarea de la red neuronal discriminadora se vuelve más difícil. Así pues, tanto el generador como el discriminador van mejorando simultáneamente.

El generador crea caras distintas. Las caras son distintas porque la entrada es distinta. Como entrada utilizamos un vector aleatorio (que llamamos Z). Usando otro vector aleatorio, tendremos otra cara … o lo que quiera que estemos generando.
